```python
# prompt: give the code to upload dataset by using url
import pandas as pd

# Define column names, assuming your data has 61 columns
names = [i for i in range(61)]

url = "/content/drive/MyDrive/Sonar.csv"
sonar_data = pd.read_csv(url, names=names)

# Access the 60th column using its index (starting from 0)
sonar_data[60].value_counts()
```

M --> Mine

R --> Rock

```python
sonar_data.groupby(60).mean()
```

```python
# seperating data and labels
X = sonar_data.drop(columns=60,axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

Training and Test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training --> Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the logistic regression model without training data
model.fit(X_train,Y_train)
```

Model Evaluation

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data : ",training_data_accuracy)
```

```python
#accuracy on training data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on test data : ",test_data_accuracy)
```
